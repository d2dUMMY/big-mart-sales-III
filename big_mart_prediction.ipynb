```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import random
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv("./dataset/Train_UWu5bXk.csv")
```

```python
test=pd.read_csv("./dataset/Test_u94Q5KV.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
test["Item_Outlet_Sales"] = np.NaN
```

```python
train["Status"] = "train"
test["Status"] = "test"
```

```python
df = pd.concat([train, test], ignore_index=True)
df.info()
```

```python
df["New_Item_Type"] = df["Item_Identifier"].str[0].map({
    'F': 'Food',
    'D': 'Drink',
    'N': 'Non Consumable'
})

df.head()
  
```

```python
df['Item_Fat_Content'].replace(to_replace ="low fat", 
                 value ="Low Fat",inplace=True) 
df['Item_Fat_Content'].replace(to_replace ="LF", 
                 value ="Low Fat",inplace=True) 
df['Item_Fat_Content'].replace(to_replace ="reg", 
                 value ="Regular",inplace=True) 
```

```python
df.Item_Fat_Content.value_counts() 
```

```python
train.Item_Visibility[train.Item_Visibility > 0].mean()
```

```python
train.Item_Visibility.mean()
```

```python
df.Item_Visibility.replace(to_replace = 0, 
                              value= df.Item_Visibility[df.Item_Visibility > 0].mean(), 
                              inplace = True)
```

```python
df.loc[df.New_Item_Type == 'Non Consumable', 'Item_Fat_Content'] = "NIL"
df.head()
```

```python
df['Outlet_Size'].fillna('Medium',inplace=True)
```

```python
mean_value=df['Item_Weight'].mean()
df['Item_Weight']=df['Item_Weight'].fillna(mean_value)
```

```python
print("Seeing number of null values in testing dataset")
print(pd.isnull(df).sum())
```

```python
df.Outlet_Establishment_Year = 2009 - df.Outlet_Establishment_Year
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
columns = ['Outlet_Establishment_Year','Item_MRP','Item_Visibility','Item_Weight']

for col in columns:
    scale = StandardScaler()
    df[col] = pd.Series(scale.fit_transform(df[col].values.reshape(-1,1)).reshape(-1,))
```

```python
from sklearn.preprocessing import LabelEncoder
columns = ['Item_Identifier', 'Item_Fat_Content', 'Item_Type','Outlet_Identifier',
           'Outlet_Size','Outlet_Location_Type','Outlet_Type','New_Item_Type']

for col in columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
train = df[df.Status == "train"]
test = df[df.Status == 'test']
```

```python
test.drop(["Item_Outlet_Sales", "Status"], axis=1,inplace=True)
```

```python
train.drop(['Status'],axis=1,inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(criterion = 'mse')

from sklearn.model_selection import GridSearchCV
parameters = {"max_depth": [5,10,15,20],
             "min_samples_split": [2,4,6,8],
             "min_samples_leaf": [2,4,6,8,10]}

clf = GridSearchCV(reg, parameters, cv=5, verbose=2)

clf.fit(train.drop(["Item_Outlet_Sales"], axis=1), train.Item_Outlet_Sales)
```

```python
clf.best_params_
```

```python
ans =clf.predict(test)
```

```python
test = pd.read_csv("./dataset/Test_u94Q5KV.csv")
```

```python
item_id = test.Item_Identifier
out_id = test.Outlet_Identifier
result = pd.DataFrame(data = {"Item_Identifier":test.Item_Identifier, 
                              "Outlet_Identifier":test.Outlet_Identifier,
                              "Item_Outlet_Sales":ans})
result.to_csv("./dataset/decision_tree.csv",index=False)
```
